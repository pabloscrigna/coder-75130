```python
#clase

class Persona:

  def __init__(self, nombre, edad):
    self.nombre = nombre
    self.edad = edad

  def saluda(self, otra_persona):
    return f'Hola {otra_persona.nombre}, me llamo {self.nombre}'



```

```python
persona = Persona("pablo", 40)
```

```python
persona.nombre
```

```python
persona.saluda(Persona("juan", 20))
```

```python
class Rectangulo:

  figura = "Rectangulo"

  def __init__(self, ladoA, ladoB):
    self.ladoA = ladoA
    self.ladoB = ladoB

  def area(self):
    return self.ladoA * self.ladoB
```

```python
rectangulo = Rectangulo(2, 3)
```

```python
rectangulo.area()
```

```python
class Cuadrado:

  def __init__(self, lado):
    self.lado = lado

  def area(self):
    return self.lado * self.lado
```

```python
cuadrado = Cuadrado(2)
```

```python
cuadrado.area()
```
